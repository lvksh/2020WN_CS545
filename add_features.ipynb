```python
import pandas as pd
```

```python
def read_features(file_name):
    df1 = pd.read_excel(file_name, sheet_name= 'Sheet1')
    df1.drop(columns = [df1.columns[0], df1.columns[1]], inplace= True)
    df1.rename(columns={'日期': '股票名称', 'Date': '股票代码'}, inplace = True)

    df2 = pd.read_excel(file_name, sheet_name= 'Sheet2')
    df2.drop(columns = [df2.columns[0], df2.columns[1]], inplace= True)
    df2.rename(columns={'日期': '股票名称', 'Date': '股票代码'}, inplace = True)

    df3 = pd.read_excel(file_name, sheet_name= 'Sheet3')
    df3.drop(columns = [df3.columns[0], df3.columns[1]], inplace= True)
    df3.rename(columns={'日期': '股票名称', 'Date': '股票代码'}, inplace = True)

    ts_code = list(df1['股票代码'])
            # df1：10日指数移动均线_PIT

    df1_ = df1.T[[0]].iloc[2:]
    df2_ = df2.T[[0]].iloc[2:]
    df3_ = df3.T[[0]].iloc[2:]
    code1 = pd.DataFrame([ts_code[0] for _ in range(df1_.shape[0])], index=df1_.index)
    df_ = pd.concat([code1, df1_, df2_, df3_], axis=1)
    df_.columns = ['ts_code', 'EMA10', 'turnoverrate10', 'volatilityratio10']
    df_.sort_index(ascending=False, inplace = True)

    for i in range(1, len(ts_code)):
        df1_cur = df1.T[[i]].iloc[2:]
        df2_cur = df2.T[[i]].iloc[2:]
        df3_cur = df3.T[[i]].iloc[2:]
        code2 = pd.DataFrame([ts_code[i] for _ in range(df1_cur.shape[0])], index=df1_cur.index)
        df_cur = pd.concat([code2, df1_cur, df2_cur, df3_cur], axis=1)
        df_cur.columns = ['ts_code', 'EMA10', 'turnoverrate10', 'volatilityratio10']
        df_cur.sort_index(ascending=False, inplace = True)
        df_ = pd.concat([df_, df_cur])
        print('Finished stock {} : {}'.format(i, ts_code[i]))
    return df_


def add_features_to_daily(daily_path, code_filename_lst):
    # code_filename_lst = ['sh600000']
    # daily_path = 'daily/'
    string_add = 'numerical_features.xlsx'
    df_add = read_features(string_add)
    
    for code_filename in code_filename_lst:
        string_daily = daily_path+'/'+code_filename+'.csv'
        daily_file = pd.read_csv(string_daily)
        
        df_feature = df_add[df_add['ts_code'] == daily_file['ts_code'][0]]
        df_feature.drop(columns = ['ts_code'], inplace= True)
        df_feature.reset_index(drop=True, inplace = True)
        df = pd.concat([daily_file , df_feature], axis=1)
    return df
    
```

```python
code_filename_lst = ['sh600000']
daily_path = 'daily/'
df = add_features_to_daily(daily_path, code_filename_lst)
outputpath = 'daily/sh600000_add_features.csv'
df.to_csv(outputpath)
```
